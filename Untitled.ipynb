```python
import plotly.graph_objects as go

data = [{'x':[1,2,3,4,5],'y':[6,7,8,9,10]},
        {'x':[2,4,6,8,10],'y':[4,8,12,16,10]}]

fig=go.Figure()
for i in range(len(data)):
  fig.add_trace(go.Scatter(x=data[i]['x'],y=data[i]['y'],mode='lines'))
fig.update_layout(title='Time Series',xaxis_title='Time',yaxis_title='Values')
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate data for the 3D scatter plot
x = np.linspace(0, 10, 100)  # X-coordinates
y = np.linspace(0, 10, 100)  # Y-coordinates
z = np.random.randn(100)     # Z-coordinates (random values)

# Create the 3D scatter plot with markers
fig = go.Figure(
    data=[go.Scatter3d(x=x,y=y,z=z,mode='markers',
          marker=dict(
            size=5,           # Size of the markers
            color=z,          # Color the markers based on z-values
            colorscale='Viridis',  # Color scale for the markers
            opacity=0.8       # Transparency of the markers
        )
    )]
)


fig.update_layout(
    title='3D Scatter Plot with Markers',scene=dict(xaxis_title='X-Axis',yaxis_title='Y-Axis',zaxis_title='Z-Axis'))

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # Sine wave
y2 = np.cos(x)  # Cosine wave

# Create the plot
plt.figure(figsize=(8, 6))

# Plot y1 with line formatting
plt.plot(x, y1, label='Sine Wave', color='blue', linestyle='--', linewidth=2, marker='o', markersize=5)

# Plot y2 with line formatting
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='-', linewidth=2, marker='s', markersize=5)

# Add title and labels
plt.title('Linear Plotting with Line Formatting', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Add a legend
plt.legend(loc='upper right', fontsize=12)

# Add grid lines
plt.grid(True, linestyle=':', linewidth=0.5)

# Show the plot
plt.show()
```

```python
# Program to analyze a sentence
def analyze_sentence(sentence):
    words = len(sentence.split())
    digits = sum(char.isdigit() for char in sentence)
    uppercase = sum(char.isupper() for char in sentence)
    lowercase = sum(char.islower() for char in sentence)
    return words, digits, uppercase, lowercase

# Accept a sentence from the user
sentence = input("Enter a sentence: ")

words, digits, uppercase, lowercase = analyze_sentence(sentence)
print(f"Number of words: {words}")
print(f"Number of digits: {digits}")
print(f"Number of uppercase letters: {uppercase}")
print(f"Number of lowercase letters: {lowercase}")
```

```python
x = input("Enter first string: ")
y = input("Enter second string: ")
x = x.strip()
y = y.strip()

sim = 0

# Determine the longer and shorter string
if len(x) > len(y):
    longer = x
    shorter = y
else:
    longer = y
    shorter = x

# Compare characters from the shorter string with the longer string
j = 0
for i in shorter:
    if i == longer[j]:
        sim += 1
    j += 1

# Calculate similarity
similarity = sim / len(longer)
print("The similarity between the two given strings is", similarity)
```

```python
# Program to compute Fibonacci sequence
def fibonacci(n):
    if n <= 0:
        raise ValueError("N must be greater than 0")
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]

try:
    n = int(input("Enter a value for N (N > 0): "))
    if n <= 0:
        raise ValueError("N must be greater than 0")
    fib_sequence = fibonacci(n)
    print(f"The first {n} Fibonacci numbers are: {fib_sequence}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def binary_to_decimal(binary):
    decimal = 0
    pow = 0
    # Process the binary string from right to left
    for digit in reversed(binary):
        decimal += int(digit) * (2 ** pow)
        pow += 1
    return decimal

def octal_to_hexadecimal(octal):
    decimal = 0
    pow = 0
    # Process the octal string from right to left
    for digit in reversed(octal):
        decimal += int(digit) * (8 ** pow)
        pow += 1

    hexadecimal = ""
    hexa_chars = '0123456789ABCDEF'
    while decimal > 0:
        remainder = decimal % 16
        hexadecimal = hexa_chars[remainder] + hexadecimal
        decimal //= 16
    return hexadecimal
```

```python
# Program to check if a number is palindrome and count digit occurrences
def is_palindrome(number):
    return str(number) == str(number)[::-1]

def count_digits(number):
    digit_count = {}
    for digit in str(number):
        digit_count[digit] = digit_count.get(digit, 0) + 1
    return digit_count

# Accept input from the user
number = int(input("Enter a number: "))

if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")

digit_count = count_digits(number)
print("Digit occurrences:")
for digit, count in digit_count.items():
    print(f"Digit {digit}: {count}")
```

```python
# Program to find the best average of two tests out of three
def best_two_average(marks):
    marks.sort(reverse=True)  # Sort marks in descending order
    return sum(marks[:2]) / 2  # Average of the best two marks

# Accept marks from the user
marks = []
for i in range(3):
    mark = float(input(f"Enter marks for Test {i + 1}: "))
    marks.append(mark)

best_avg = best_two_average(marks)
print(f"The best average of two tests is: {best_avg}")
```
